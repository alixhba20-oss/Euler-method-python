```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the ODE dy/dx = -2*x*y
def f(x, y):
    return -2 * x * y

# Step 2: Exact solution (true answer)
def y_exact(x):
    return np.exp(-x**2)

# Step 3: Euler method implementation
def euler_method(f, x0, y0, h, x_end):
    num_steps = int((x_end - x0) / h) + 1
    x_values = np.linspace(x0, x_end, num_steps)
    y_values = np.zeros(num_steps)
    y_values[0] = y0

    for i in range(1, num_steps):
        y_values[i] = y_values[i-1] + h * f(x_values[i-1], y_values[i-1])

    return x_values, y_values

# Step 4: Parameters
x0 = 0        # starting x
y0 = 1        # starting y (jab x=0 to y=1)
x_end = 2     # ending x
h = 0.1       # step size (chhota step = accurate)

# Step 5: Run Euler method
x_euler, y_euler = euler_method(f, x0, y0, h, x_end)

# Step 6: Calculate exact solution for comparison
x_exact = np.linspace(x0, x_end, 100)
y_exact_vals = y_exact(x_exact)

# Step 7: Plot graph
plt.figure(figsize=(8, 5))
plt.plot(x_euler, y_euler, 'bo-', label='Euler Method (h=0.1)', markersize=4)
plt.plot(x_exact, y_exact_vals, 'r-', linewidth=2, label='Exact Solution')
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Euler Method vs Exact Solution', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()

# Step 8: Print results
print("Results at x = 2:")
print(f"Euler method: {y_euler[-1]:.4f}")
print(f"Exact solution: {y_exact(2):.4f}")
print(f"Error: {abs(y_euler[-1] - y_exact(2)):.4f}")
```
